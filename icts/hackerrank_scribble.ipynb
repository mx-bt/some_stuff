```python
class EvenStream(object):
    def __init__(self):
        self.current = 0

    def get_next(self):
        to_return = self.current
        self.current += 2
        return to_return

class OddStream(object):
    def __init__(self):
        self.current = 1

    def get_next(self):
        to_return = self.current
        self.current += 2
        return to_return


def print_from_stream(n, stream=None): # to debug
    if stream == None:
        stream = EvenStream()
    for _ in range(n):
        print(stream.get_next())


# queries = int(3)
# samples = ["odd 2","even 3","odd 5"]
for i in range(int(len(test_input)/2)):
    stream_name, n = test_input[i*2],test_input[(2*i)+1]
    n = int(n)
    print("Command: ",stream_name," ",n)
    if stream_name == "even":
        print_from_stream(n)
    else:
        print_from_stream(n, OddStream())
```
