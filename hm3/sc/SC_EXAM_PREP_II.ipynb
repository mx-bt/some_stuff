```python
#   Exercise 5.3: Taylor Series, sympy and Documentation
#   polynomial of 4th degree for y=f(x)=sin(x) at x0 = 0 (McL)
import numpy as np
import sympy as sp

x_sy = sp.symbols("x")

def f(x):
    return sp.sin(x)

def taylor(f,x0=1,N=1):
    taylor_approxmation = 0
    for n in range(N+1):
        taylor_approxmation += sp.diff(f(x),x_sy,n).subs(x_sy, x0)/sp.factorial(n)*(x_sy-x0)**n
    return taylor_approxmation

f_T = taylor(f,0,10)
display(f_T)
f_T_n = sp.lambdify(x_sy,f_T,"numpy")

import matplotlib.pyplot as plt

x_min = -np.pi*5
x_max = np.pi*5

x_num = np.linspace(x_min, x_max, 1000)
y1 = np.sin(x_num)
y2 = f_T_n(x_num)
plt.plot(x_num, y1, label='Base function f(x)', color='blue')
plt.plot(x_num, y2, label='Taylor approximation', color='red')
plt.xlabel('x')
plt.xlim(x_min,x_max)
plt.ylabel('y')
plt.ylim(-2,2)
plt.legend()
plt.show()
```

```python
#   Exercise 5.7: Count Occurrences of a String in a String
#   count_substrings.py

gene = "AGTCAATGGAATAGGCCAAGCGAATATTTGGGAAAAAAAAAAAAAAAATATATTATATTAGGGCTACCAAGTCAATGGAATAGGCCAAGCGAATATTTGGCTACCA"
import re
#   GAATAGG
#   GAATATTTGG

def freq(letter, text):
    count = 0
    for lookup in text:
        if lookup == letter:
            count += 1
        else:
            pass
    return count/len(text)

print("my function: ",freq("G", gene))
print("python tool: ",gene.count("G")/len(gene))
     

def pairs(letter, text):
    count = 0
    for i in range(len(text)-1):
        if (text[i] == text[i+1])and(text[i] == letter) and (text[i] != text[i-1]):
            count += 1
        else:
            pass
    return count

print("my function: ", pairs("G",gene))
print("python tool: ", gene.count("G"*2))

def mystruct(text):
    count = 0
    for i in range(len(text)-1):
        if text[i]=="G":
            for j_ in range(len(text)-2-i):
                j = j_+1+i
                if text[j]=="A" or text[j]=="T":
                    pass
                else:
                    if text[j]=="G" and text[j+1]=="G" and text[j-1]!="G":
                        count += 1
                        break
                    else:
                        break
    return count

print("my function: ",mystruct(gene))
print("python tool: ", len(re.findall("G[AT]+?GG", gene)))
```

```python
#   Exercise 6.1: Hand Calculations for the Trapezoidal Method




# trapezoid with hand

import math as m
def trapezoidal(f,a,b,n):
    integral_value = 0
    area = 0
    interval = b-a
    t_x = interval/n
    for i in range(n):
        p1 = a+(i*t_x)
        p2 = a+((i+1)*t_x)
        a_ = f(p1)
        b_ = f(p2)
        h = p2-p1
        A = (a_+b_)/2*h
        integral_value += A
        area += abs(A)
    return integral_value, area

def midpoint(f,a,b,n):
    integral_value = 0
    area = 0
    interval = b-a
    t_x = interval/n
    for i in range(n):
        p1 = a+(i*t_x)
        p2 = a+((i+1)*t_x)
        A = f((p1+p2)/2)*(t_x)
        integral_value += A
        area += abs(A)

    return integral_value, area

def precise_integration(f,a,b):
    import sympy as sp
    x_sy = sp.symbols("x_sy")
    sy_int = sp.simplify(sp.integrate(f(x_sy),x_sy))
    precise = sp.lambdify(x_sy,sy_int,"numpy")
    prc = precise(b)-precise(a)
    return prc
```
