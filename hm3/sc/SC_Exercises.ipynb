```python
"""1.2"""
import matplotlib.pyplot as plt
import numpy as np

def scalarprod(vector1,vector2):
    if np.shape(vector1) != np.shape(vector2):
        print("Vectors have different shape, op not poss")
    else:
        for i in range(np.shape(vector1)[0]):
            scalproduct = vector1[i] * vector2[i]
    return scalproduct

"""1.2 - 2 """
# for two 1,3 vectors
# general !?

def crossp(vector1,vector2):
    returnvector = np.zeros_like(vector1)
    if (np.shape(vector1) != np.shape(vector2)) or np.shape(vector1)[0] != 3:
        print("Error, the vectors don't have the equal size or are not of shape (3,1)")
        pass
    else:
        returnvector[0] = vector1[1]*vector2[2] - vector1[2]*vector2[1]
        returnvector[1] = vector1[2]*vector2[0] - vector1[0]*vector2[2]
        returnvector[2] = vector1[0]*vector2[1] - vector1[1]*vector2[0]
    print("Cross Product Vector= ", returnvector)
    return returnvector

def crossp_plot(vector1,vector2):
    from mpl_toolkits.mplot3d import Axes3D
    import matplotlib.pyplot as plt
    vector3 = crossp(vector1,vector2)
    U = np.array([vector1[0],vector2[0],vector3[0]])
    V = np.array([vector1[1],vector2[1],vector3[1]])
    W = np.array([vector1[2],vector2[2],vector3[2]])

    X = np.array([0,0,0])
    Y = np.array([0,0,0])
    Z = np.array([0,0,0])

    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.quiver(X, Y, Z, U, V, W)

    ax.set_xlim([np.min(U), np.max(U)])
    ax.set_ylim([np.min(V), np.max(V)])
    ax.set_zlim([np.min(W), np.max(W)])

    return plt.show()

def scalartripleproduct(vectorA, vectorB, vectorC):
    # in DE = Spatprodukt
    crossproductvector = crossp(vectorA,vectorB)
    scalarproductR = scalarprod(crossproductvector,vectorC)
    return scalarproductR
```

```python
"""MONTE CARLO INTEGRATION"""

#functon fofx
def f_of_x(x):
    return np.log10(x) * np.sin(10*x) + 1

def mci(xintmin=0,xintmax=1,precision=100):
    import random as rd
    xtest = np.linspace(xintmin,xintmax,precision)
    ymin = np.min(f_of_x(xtest))-1
    ymax = np.max(f_of_x(xtest))+1

    #random points
    xspace = np.zeros(precision)
    yspace = np.zeros(precision)

    totalarea = (ymin-ymax)*(xintmin-xintmax)

    for i in range(precision):
        xspace[i] = rd.uniform(xintmin,xintmax)
        yspace[i] = rd.uniform(ymin,ymax)
    
    integrationcount = 0
    for index in range(len(yspace)):
        if yspace[index] >= 0:
            if f_of_x(xspace[index]) >= yspace[index]:
                integrationcount += 1
            else:
                pass
        if yspace[index] <= 0:
            if f_of_x(xspace[index]) <= yspace[index]:
                integrationcount += 1
            else:
                pass
    area = ((integrationcount/precision))*totalarea
    integrationcount = 0
    return area

mci(1,10,100000)
```

```python
# Creating random matrices to truly test the versatility of our code
def createrandomfloatmatrix(m_shape,n_shape):
    import random as rd
    matrix = np.zeros((m_shape, n_shape))
    for m in range(m_shape):
        for n in range(n_shape):
            matrix[m,n] = rd.randrange(1,10)
        
    return matrix

def triangular(A,b):
    for m in range(np.shape(A)[0]):
        # print(m) 0,1,2
        f = A[m,m]
        A[m,:] =  A[m,:] / f
        b[0,m] =  b[0,m] / f

        for i in range(np.shape(A)[0]-1-m):
            A[i+1+m,:] = A[i+1+m,:] - A[m,:]*A[i+1+m,m]
            b[0,i+1+m] = b[0,i+1+m] - b[0,m]*A[i+1+m,m]
    return A,b

# backsub !?




def solveAxb(A,b):
    # starting from row m(max)-1, subract this row * value of target row from all other rows
    # ... repeat until target row indices iterated
    # ... repeat until "killer" row is row index 1
    A,b = triangular(A,b)

    for tm in range(np.shape(A)[0]-1):
        i = np.shape(A)[0] - tm - 1
        for m in range(np.shape(A)[0]-1-tm):
            A[m,:] = A[m,:] - A[i,:]*A[m,i]
            b[0,m] = b[0,m] - b[0,i]*A[m,i]
    return A,b

mA = createrandomfloatmatrix(5,5)
vb = createrandomfloatmatrix(1,5)
print("Matrix Org")
print(mA)
print("Vector Org")
print(vb)

rmA, rvb = solveAxb(mA,vb)

print("Result Matrix: ")
print(np.round(rmA))
print("Result Vector: ")
print(np.round(np.transpose(rvb)))
```

```python
"""MONTE CARLO INTEGRATION ON STEROIDS"""

import random
import numpy as np
import time 
import matplotlib.pyplot as plt

#Function
def f(x):
    return np.cos(x) ** 2 + np. sin(2*x)

# enter notorious function here
#while True:
#    try:
#        user_input = np.sin(x) # input("Enter function f(x): ")
#        f = eval(f"lambda x: {user_input}")
#        break
#    except:
#        print("Error: Invalid input. Please enter a valid function.")

# Adjust measurement area (X Values)
while True:
    try:
        user_inputx1 = -5 # input("Enter your Xmin: ")
        test_area_x_start = float(user_inputx1)
        break
    except ValueError:
        print("Error: Invalid input. Please enter a number.")
while True:
    try:
        user_inputy1 = 5 # input("Enter your Xmax: ")
        test_area_x_end = float(user_inputy1)
        break
    except ValueError:
        print("Error: Invalid input. Please enter a number.")
range_x = test_area_x_end - test_area_x_start

# Adjust precision
while True:
    try:
        user_inputy1 = 100000 # input("Enter the number of desired test points: ")
        no_sample_points = int(user_inputy1)
        if no_sample_points > 100000000:
            raise ValueError("Error: Input must be less than or equal to 100000000.")
        break
    except ValueError:
        print("Error: Invalid input. Please enter a number.")

# We create the function values for our desired sample resolution
list_x = np.linspace(test_area_x_start, test_area_x_end, no_sample_points)
list_y = f(list_x)

# Boundaries for the y values
test_area_y_start = np.min(list_y)
test_area_y_end = np.max(list_y)
range_y = test_area_y_end - test_area_y_start

# Create random points in the 
random_x_values = np.array([random.uniform(test_area_x_start, test_area_x_end) for i in range(no_sample_points)])
random_y_values = np.array([random.uniform(test_area_y_start, test_area_y_end) for i in range(no_sample_points)])
random_points = []

for xi in (range(len(random_x_values))):
    random_points.append([random_x_values[xi],random_y_values[xi]])

# Evaluating the number of points that lay within the function
area = 0
not_area = 0
color_label = []

for i in (range(len(random_points))):
    if (random_points[i][1] >= 0) & (random_points[i][1] <= f(random_points[i][0])):
        area += 1
        random_points[i].append("pos")
        color_label.append("red")
    elif (random_points[i][1] <= 0) & (random_points[i][1] >= f(random_points[i][0])):
        area -= 1
        random_points[i].append("neg")
        color_label.append("red")
    else:
        not_area += 1
        random_points[i].append("nul")
        color_label.append("blue")

approx_area = np.round((area / no_sample_points)*(range_x * range_y),4)

print("The area of the approximated integral is:",
      approx_area
)

# Visualizing the result

print(random_points[:3][:5])
print(random_points[1][0])
print(random_points[0][1])

#print(random_points[:10][1])
#print(random_points[:10][2])

#Plotting
fig, ax = plt.subplots()
ax.scatter(random_x_values, random_y_values, c=color_label, label='Area between f(x) and x-axis')
ax.plot(list_x, list_y, c="red",label='Function f(x)')
ax.set_title(f'Monte Carlo Area Approximation = {approx_area}')
ax.legend()
plt.show()
```
