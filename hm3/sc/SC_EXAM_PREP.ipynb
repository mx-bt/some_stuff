```python
#   Exercise 3.11: Compute π
#   compute_pi.py

import numpy as np
import matplotlib.pyplot as plt

N = 100
Ns = np.linspace(1, N, N, dtype=int)
def pi_series(n):
    srs = []
    for pi in range(n):
        srs.append(np.pi)
    return srs

true_pi_series = pi_series(N)


def Leibniz_Pi(n):
    l = []
    leibniz_pi = 0
    for k in range(n):
        leibniz_pi += 8/((4*k+1)*(4*k+3))
        l.append(leibniz_pi)
    return leibniz_pi, l

print("Leibniz", Leibniz_Pi(N)[0])
leibniz_series = Leibniz_Pi(N)[1]

def Euler_Pi(n):
    e = []
    euler_pi = 0
    for k in range(n):
        euler_pi += 6/((k+1)**2)
        e.append(np.sqrt(euler_pi))
    euler_pi = np.sqrt(euler_pi)
    return euler_pi, e


print("Euler: ",Euler_Pi(N)[0])
euler_series = Euler_Pi(N)[1]


l_error = []
e_error = []
for k in range(N):
    e_error.append(abs(Euler_Pi(N)[1][k]-np.pi)/np.pi)
    l_error.append(abs(Leibniz_Pi(N)[1][k]-np.pi)/np.pi)
        

fig, ax1 = plt.subplots()
ax2 = ax1.twinx()

ax1.plot(true_pi_series, label="True Pi Series")
ax1.plot(leibniz_series, label="Leibniz Series")
ax1.plot(euler_series, label="Euler Series")

ax2.plot(l_error, label="Leibniz Error",color="red")
ax2.plot(e_error, label="Euler Error",color="red")
ax2.set_ylabel('Relative Error',color="red")

ax1.set_xlabel("N")
ax1.set_ylabel("π Approximation")

ax1.legend()
plt.tight_layout()
plt.show()
```

```python
#   Exercise 4.12: Fit Straight Line to Data
#    fit_straight_line.py
import numpy as np
import matplotlib.pyplot as plt
fig, ax1 = plt.subplots()

yis = [0.5,
       2.0,
       1.0,
       1.5,
       7.5
      ]
xis = np.linspace(0,4,len(yis))

a = 5 #int(input("Input initial value for a(int): "))
b = -10 #int(input("Input initial value for b(int): "))
search_step = 0.02 #float(input("Input the size of each linear search step: "))
epsilon = 0.0000001 #float(input("Desired precision(float): "))

def line(x,a,b):
    return(a*x+b)
def e(yis,xis,a,b):
    if len(yis) != len(xis):
        print("Yi and Xi arrays aren't equal")
        pass
    sum_e = 0
    for i in range(len(yis)):
        sum_e += (line(xis[i],a,b) - yis[i])**2
    return sum_e

errors = [e(yis,xis,a,b)]
print("Initial rror: ",errors)
cnt = 0
delta = np.inf
while (delta > epsilon) and cnt <= 1000:
    cnt += 1
    a_try1 = a+search_step
    a_try2 = a-search_step
    if e(yis,xis,a_try2,b) < e(yis,xis,a_try1,b):
        candidate = a_try2
    else:
        candidate = a_try1
    if e(yis,xis,candidate,b) < errors[cnt-1]:
        a = candidate
    else:
        pass

    b_try1 = b+search_step
    b_try2 = b-search_step
    if e(yis,xis,a,b_try2) < e(yis,xis,a,b_try1):
        candidate = b_try2
    else:
        candidate = b_try1
    if e(yis,xis,a,candidate) < errors[cnt-1]:
        b = candidate
    else:
        pass
    errors.append(e(yis,xis,a,b))
    print("Error: ",errors[cnt])
    delta = np.abs(errors[cnt-1]-errors[cnt])

    if cnt == 1000:
        print("No convergence after 1000 iterations")
        print(f"The best parameters are: a={a}, b={b}")
        print(f"Initial error= {errors[1]}")
        print(f"Lowest error= {errors[cnt]}")
    else:
        ("Fail pass")
        pass

if delta < epsilon:
    print(f"Convergence after {cnt} iterations")
    print(f"The best parameters are: a={a}, b={b}")
    print(f"Initial error= {errors[1]}")
    print(f"Lowest error= {errors[cnt]}")

line_plot = []
for i in range(len(xis)):
    line_plot.append(line(xis[i],a,b))
ax1.plot(xis,line_plot,label="predictions")


plt.title(f"Final error={e(yis,xis,a,b)} after {cnt} iterations")
ax2 = ax1.twinx()
ax1.scatter(xis,yis, label="yi")
ax1.set_xlabel("x")
ax1.set_ylabel("y=f(x)")
ax1.legend()
plt.tight_layout()
plt.show()
```

```python
#   Exercise 4.13: Fit Sines to Straight Line
#   fit_sines.py
import numpy as np
import matplotlib.pyplot as plt

def f(t):
    return (1/np.pi)*t

def g(t):
    return 0.1*t+0

def sinesum(t,b):
    N = len(b)
    S_N = 0
    for n in range(N):
        S_N += b[n]*np.sin((n+1)*t)
    return S_N

t_test = np.array([-np.pi/2,np.pi/4])
b_test = np.array([4,-3])
print(sinesum(t_test,b_test))

t = np.linspace(-np.pi,np.pi,100)

b_length = 10
b = np.ones_like(range(b_length))

def plot_compare(f,N,M,show=1):

    fig, ax1 = plt.subplots()
    t = np.linspace(-np.pi,np.pi,M) # M is the number of uniformly distributed t coordinates used to plot f and SN
    b = np.ones_like(range(N)) # N is the number of terms in the sum SN (t)
    plt.title(f"Plot of f(t) and SN(t)")
    ax1.plot(t,sinesum(t,b), label="sinesum(t,b)")

    if show==1:
        ax1.plot(t,f(t),label="f(t)")
        ax1.set_xlabel("x")
        ax1.set_ylabel("y=f(x)")
        ax1.legend()
        plt.tight_layout()
        plt.show()
    else:
        pass

def error(b,f,M=100):
    E = 0
    t = np.linspace(-np.pi,np.pi,M)
    for m in t:
        E += (f(m)-sinesum(m,b))**2
    E = np.sqrt(E)
    return E

def trial_auto(f,N,tries=1):
    fig, ax1 = plt.subplots()
    M = 500
    for a_try in range(tries):
        b=[]
        for j in range(N):
            b.append(np.random.random()*10)
        ax1.plot(t,sinesum(t,b), label=f"try no.{a_try+1};error={error(b,f,M)}")
        #print(error(b,f,M))

        if a_try+1 == tries:
            ax1.plot(t,f(t),label="f(t)")
            ax1.set_xlabel("x")
            ax1.set_ylabel("y=f(x)")
            ax1.legend()
            plt.ylim(-10,10)
            plt.tight_layout()
            plt.show()    


def trial_man(f,b):
    fig, ax1 = plt.subplots()
    M = 500

    ax1.plot(t,sinesum(t,b), label=f"approximation")
    ax1.plot(t,f(t),label="f(t)")
    ax1.set_xlabel("x")
    ax1.set_ylabel("y=f(x)")
    ax1.legend()

    plt.title(f"Error={error(b,f,M)}")
    #plt.ylim(-10,10)
    plt.tight_layout()
    plt.show() 


def autosearch(f):
    best_bs = []
    iteration = 0
    best_error = np.inf
    for b1 in np.linspace(-1,1,201):
        for b2 in np.linspace(-1,1,201):
            for b3 in np.linspace(-1,1,201):
                iteration += 1
                
                b_exp = [b1,b2,b3]
                temp_error = error(b_exp,f)
                print(f"Iteration={iteration}, Error={temp_error}")

                if temp_error < best_error:
                    best_error = temp_error
                    best_bs = b_exp
                    print(f"Iteration={iteration}, New_best_error={temp_error}")
                else:
                    pass
    print(f"Best b's:{best_bs}, best error:{best_error}")
    trial_man(f,best_bs)

autosearch(f)
```
