```python
from joblib import load
import numpy as np

mnist = load("C:/Users/MaxB2/Documents/Machine_Is_Learning/mnist_dataset_784_v1.joblib")

X,y = mnist["data"],mnist["target"]
y = y.astype(np.uint8)
```

```python
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt

tsne = TSNE(n_components=2, perplexity=30, random_state=42)
X_tsne = tsne.fit_transform(X)
```

```python
tsne_3d = TSNE(n_components=3, perplexity=30, random_state=42)
X_tsne_3d = tsne_3d.fit_transform(X)
```

```python
colors = plt.cm.rainbow(y / 10.0)
plt.scatter(X_tsne[:, 0], X_tsne[:, 1],c=colors,alpha=0.1)
plt.xlabel('t-SNE Dimension 1')
plt.ylabel('t-SNE Dimension 2')
plt.title('t-SNE Visualization')
plt.colorbar()
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
colors = plt.cm.rainbow(y / 10.0)
ax.scatter(X_tsne_3d[:, 0], X_tsne_3d[:, 1], X_tsne_3d[:,2], c=colors, marker='o',alpha=0.1)

%matplotlib qt
ax.set_xlabel('t-SNE Dimension 1')
ax.set_ylabel('t-SNE Dimension 2')
ax.set_zlabel('t-SNE Dimension 3')
plt.show()
```
